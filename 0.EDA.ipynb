```python
import chardet
with open('dataset/2023년_월별_행정동별_10개관심집단수.csv', 'rb') as file:
    result = chardet.detect(file.read())
    encoding = result['encoding']
```

```python
import pandas as pd
df = pd.read_csv('dataset/2023년_월별_행정동별_10개관심집단수.csv', encoding=encoding)
df.head()
```

```python
df.shape
df.describe()
df.info()
```

```python
for column in df.columns:
    print(column)
    print(df[column].unique())
    print('---------')
```

```python
# 월별 lineplot
```

```python
y_labels = list(df.columns[5:-2])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for y in y_labels:
    sns.lineplot(x=df['월'], y=df[y], hue=df['성별'])
    plt.show()
```
