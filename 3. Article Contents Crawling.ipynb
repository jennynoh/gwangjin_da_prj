```python
import pandas as pd 
df = pd.read_csv("dataset/광진구_news_2023.01.01_2023.03.31_1분기.csv")
df.head()
```

```python
# 월별 등장단어 빈도 그래프 반환 함수
def word_freq(df, month):
    month_df = df[(pd.to_datetime('2023.01.01') <= df['pub_date']) & (df['pub_date'] < pd.to_datetime('2023.02.01'))]
```

```python
df['pub_date'] = df['pub_date'].map(lambda x : pd.to_datetime(x[:-1]))
```

```python
jan = df[(pd.to_datetime('2023.01.01') <= df['pub_date']) & (df['pub_date'] < pd.to_datetime('2023.02.01'))]
jan
```

```python
jan_titles = jan['title'].tolist()
jan_title_txt = ' '.join(jan_titles)
```

```python
jan_title_txt
```

```python
from konlpy.tag import Okt

tokenizer = Okt()
raw_pos_tagged = tokenizer.pos(jan_title_txt, norm=True, stem=True)
print(raw_pos_tagged)
```

```python
cleansed_words = []
for word in raw_pos_tagged:
    if ((word[1] == 'Noun') & (len(word[0])!=1)):
        cleansed_words.append(word[0])

print(cleansed_words)
```

```python
from collections import Counter
word_dic = dict(Counter(cleansed_words))
print(word_dic)
```

```python
sorted_dic = sorted(word_dic.items(), key=lambda x: x[1], reverse=True)
print(sorted_dic)
```

```python
# 단어 등장 빈도 시각화 
import nltk
import matplotlib.pyplot as plt

# 한글폰트 설정
plt.rcParams['font.family'] = 'AppleGothic'

# FreqDist 객체 생성
word_freq = nltk.FreqDist(cleansed_words)

# 빈도수 상위 20개 단어 선택
top_words = word_freq.most_common(20)

# 시각화
plt.figure(figsize=(15,7))
word_freq.plot(20)
plt.show()
```
