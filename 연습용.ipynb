```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc("font", family = "Malgun Gothic")
sns.set(font="Malgun Gothic", 
rc={"axes.unicode_minus":False}, style='white')
import os
import requests
import re
import json
```

```python
# 인증키 가져오기
with open('인증키.txt','r', encoding='utf-8') as f:
    key=f.read()
# 정규 표현식 패턴 정의
pattern = r"\b[a-f0-9]+\b"

# re.findall()을 사용하여 텍스트에서 패턴의 모든 발생을 찾습니다.
matches = re.findall(pattern, key)
auth_key=''
# 일치 항목 출력
for match in matches:
    auth_key+=match
# print(auth_key)
```

```python
# url 설정하기
start_index=1
end_index=1000
url= f"http://openAPI.gwangjin.go.kr:8088/{auth_key}/json/GwangjinModelRestaurantDesignate/{start_index}/{end_index}/"
```

```python
# api 로 데이터 추출 및 json으로 수정
res=requests.get(url)
res_json=res.json()
# print(res)
# print(res.json())
res_raw_json=res.json().get('GwangjinModelRestaurantDesignate').get('row')
```

```python
# 확인된 json 데이터 내용 확인

print(res_raw_json[0])
print(res_raw_json[130])
```
